```python
import torch
from matplotlib import pyplot as plt
```

```python
def rosenbrock(x, y):
    value = (1 - x) ** 2 + 100 * (y - x ** 2) ** 2
    return value.mean()
```

```python
def f(x):
    return rosenbrock(x, torch.ones(2))
```

```python
params = torch.tensor([50., 50.])
params.requires_grad_()

print(f(params))
# Do gradient descent
n_optim_steps = int(5e4)
optimizer = torch.optim.LBFGS([params], lr=1e-2)

X, Y = [], []

for ii in range(n_optim_steps):
    optimizer.zero_grad()
    loss = f(params)
    
    def closure():
        return f(params)
    

    if ii % 20 == 0: print('Step # {}, loss: {}'.format(ii, loss.item()))
    loss.backward()
    # Access gradient if necessary
    grad = params.grad.data
    optimizer.step(closure)
    X.append(ii)
    Y.append(loss.item())
```

```python
plt.plot(X, Y)
plt.yscale('log')
# plt.xscale('log')
plt.xlabel('Number of iteration')
plt.ylabel('Loss')
plt.show()
```
